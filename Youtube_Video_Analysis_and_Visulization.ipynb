```python
import yt_dlp
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("darkgrid")
```

```python
# ‚úÖ Input: YouTube Video URLs
video_1 = {"url": "https://www.youtube.com/watch?v=LS1zjr1wog4", "channel": "Code With Harry"}
video_2 = {"url": "https://youtu.be/i2OOPAvCDK0?si=jL_x6rPW2InG19FQ", "channel": "Learn Code"}
```

```python
# ‚úÖ Function to Extract YouTube Video Stats
def get_youtube_video_stats(video):
    try:
        ydl_opts = {"quiet": True, "skip_download": True}
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(video["url"], download=False)

        # ‚úÖ Extract Data
        stats = {
            "title": info.get("title", "N/A"),
            "channel": info.get("uploader", "N/A"),
            "views": info.get("view_count", 0),
            "likes": info.get("like_count", 0),
            "comments": info.get("comment_count", 0)
        }

        print("=" * 50)
        print("üìå Title:", stats["title"])
        print("üìå Channel:", stats["channel"])
        print("üìå Video URL:", video["url"])
        print("üëÄ Views:", stats["views"])
        print("üëç Likes:", stats["likes"])
        print("üí¨ Comments:", stats["comments"])
        print("=" * 50)

        return stats
    except Exception as e:
        print("‚ùå Error:", e)
        return None
```

```python
# ‚úÖ Get Stats for Both Videos
stats_1 = get_youtube_video_stats(video_1)
stats_2 = get_youtube_video_stats(video_2)
```

```python
# ‚úÖ Store Data
if stats_1 and stats_2:
    channels = [stats_1["channel"], stats_2["channel"]]
    views = [stats_1["views"], stats_2["views"]]
    likes = [stats_1["likes"], stats_2["likes"]]
    comments = [stats_1["comments"], stats_2["comments"]]
```

```python
# ‚úÖ (1) Bar Chart - Views Comparison
plt.figure(figsize=(6, 4))
sns.barplot(x=channels, y=views, hue=channels, palette="Blues_r", legend=False)
plt.xlabel("Channels")
plt.ylabel("Views")
plt.title("YouTube Views Comparison")
plt.show()
```

```python
# ‚úÖ (2) Pie Chart - Likes Comparison
plt.figure(figsize=(6, 4))
plt.pie(likes, labels=channels, autopct="%1.1f%%", colors=["green", "lightgreen"], startangle=90, shadow=True)
plt.title("YouTube Likes Share")
plt.show()
```

```python
# ‚úÖ (3) Line Chart - Comments Comparison
plt.figure(figsize=(6, 4))
plt.plot(channels, comments, marker="o", linestyle="-", color="red", markersize=8, linewidth=2)
plt.xlabel("Channels")
plt.ylabel("Comments")
plt.title("YouTube Comments Trend")
plt.show()
```
